```python
import pandas as pd
import sqlalchemy

from sqlalchemy import create_engine, MetaData, Table, inspect

# Criando um banco de dados na memoria local
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'

dados_clientes = pd.read_csv(url)

# Analise dos dados
dados_clientes.head()

# Salvando a tabela no banco de dados
dados_clientes.to_sql('clientes_banco', engine, index=False)
```

```python
# Verificando se a tabela esta no banco de dados
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# Fazendo uma pesquisa no banco de dados, selecionando os clientes empregados
query = 'SELECT * FROM clientes_banco WHERE Categoria_de_renda == "Empregado" '
data = pd.read_sql(query,con=engine)
data

# Salvando a tabela
data.to_sql('clientes_empregados', engine, index=False)

# Leitura de uma tabela do banco de dados
pd.read_sql_table('clientes_empregados', engine,columns=['ID_Cliente', 'Idade', 'Categoria_de_renda', 'Rendimento_anual'])
```
